# Approach.1

```java
//  Solution using Array sort.
class Solution {

  public int largestPerimeter(int[] nums) {
    Arrays.sort(nums);
    int maxPerimeter = 0;

    for (int i = 2; i < nums.length; i++) {
      if ((nums[i - 1] + nums[i - 2]) > nums[i]) {
        int perimeter = nums[i - 2] + nums[i - 1] + nums[i];
        if (perimeter > maxPerimeter) maxPerimeter = perimeter;
      }
    }

    return maxPerimeter;
  }
}
```

```java
// Best solution. 4ms
class Solution {

  public int largestPerimeter(int[] nums) {
    int counter = 1;
    int n = nums.length - 1;
    boolean isSorted = false;

    int side1 = 0;
    int side2 = 0;
    int side3 = 0;

    while (n >= 0) {
      if (!isSorted && n > 0) {
        isSorted = maxSwap(nums, n--);
      } else {
        n--;
      }

      if (counter == 1) {
        side3 = nums[n + 1];
      } else if (counter == 2) {
        side2 = nums[n + 1];
      } else if (counter >= 3) {
        side1 = nums[n + 1];
        side2 = nums[n + 2];
        side3 = nums[n + 3];
      }

      if (side1 + side2 > side3) {
        return side1 + side2 + side3;
      }

      counter++;
    }
    return 0;
  }

  public boolean maxSwap(int[] nums, int end) {
    boolean isSorted = true;
    int maxIndex = end;

    for (int i = 0; i < end; i++) {
      if (nums[i] > nums[maxIndex]) maxIndex = i;
      if (nums[i] > nums[i + 1]) isSorted = false;
    }

    if (maxIndex != end) swap(nums, maxIndex, end);
    return isSorted;
  }

  public int maxIndex(int[] nums, int end) {
    int max = 0;
    for (int i = 0; i <= end; i++) {
      if (nums[max] < nums[i]) max = i;
    }
    return max;
  }

  public void swap(int[] nums, int l, int r) {
    int temp = nums[l];
    nums[l] = nums[r];
    nums[r] = temp;
  }
}
```
