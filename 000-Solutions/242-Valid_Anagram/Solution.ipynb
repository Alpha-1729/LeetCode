```java
class Solution {

  public boolean isAnagram(String s, String t) {
    if (s.length() != t.length()) {
      return false;
    }
    Map<Character, Integer> charMap = new HashMap<>();
    char charOne, charTwo;
    int charOneFreq, charTwoFreq;
    for (int i = 0; i < s.length(); i++) {
      charOne = s.charAt(i);
      charTwo = t.charAt(i);

      if (charOne != charTwo) {
        charOneFreq = charMap.getOrDefault(charOne, 0) + 1;
        charTwoFreq = charMap.getOrDefault(charTwo, 0) - 1;
        if (charOneFreq == 0) {
          charMap.remove(charOne);
        } else {
          charMap.put(charOne, charOneFreq);
        }
        if (charTwoFreq == 0) {
          charMap.remove(charTwo);
        } else {
          charMap.put(charTwo, charTwoFreq);
        }
      }
    }
    return charMap.isEmpty();
  }
}
```

```java
class Solution {

  public boolean isAnagram(String s, String t) {
    if (s.length() != t.length()) {
      return false;
    }
    Map<Character, Integer> charMap = new HashMap<>();
    char charOne, charTwo;
    int charOneFreq, charTwoFreq;
    for (int i = 0; i < s.length(); i++) {
      charOne = s.charAt(i);
      charTwo = t.charAt(i);

      if (charOne != charTwo) {
        charMap.put(charOne, charMap.getOrDefault(charOne, 0) + 1);
        charMap.put(charTwo, charMap.getOrDefault(charTwo, 0) - 1);
      }
    }
    for (int j : charMap.values()) {
      if (j != 0) {
        return false;
      }
    }
    return true;
  }
}
```

```java
// Efficient solution.
class Solution {

  public boolean isAnagram(String s, String t) {
    if (s.length() != t.length()) {
      return false;
    }
    int map[] = new int[26];
    for (char x : s.toCharArray()) {
      map[x - 'a']++;
    }
    for (char x : t.toCharArray()) {
      map[x - 'a']--;
    }
    for (int i : map) {
      if (i != 0) {
        return false;
      }
    }
    return true;
  }
}
```
